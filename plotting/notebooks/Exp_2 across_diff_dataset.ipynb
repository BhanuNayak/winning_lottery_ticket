```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load data from Excel file
xls = pd.ExcelFile("/kaggle/input/final-plots/Experiment 2 VGG-2.xlsx")
```

```python
df1 = pd.read_excel(xls, sheet_name="VGG_cifar10", index_col=0)
df2 = pd.read_excel(xls, sheet_name="VGG_cifar100", index_col=0)
df3 = pd.read_excel(xls, sheet_name="VGG_SVHN", index_col=0)
```

```python
df1 = df1.rename(columns = {"Unnamed: 1":"% Pruned", "Accuracy on CIFAR10":"CIFAR10", "Unnamed: 3":"CIFAR100", "Unnamed: 4":"Random"})
df1 = df1.iloc[1:]
df1.dropna(inplace = True)
# Select specific columns for plotting
idx_vals = np.float64(df1["% Pruned"].values)
x_values = idx_vals
y_values = np.float64(df1["CIFAR10"].values)

# Plotting the graph
plt.plot(x_values, y_values,label='Cifar 10', marker = '.')
y_values = np.float64(df1["CIFAR100"].values)
plt.plot(x_values, y_values,label='Cifar 100', marker = '^')
y_values = np.float64(df1["Random"].values)
plt.plot(x_values, y_values,label='Random', marker = '*')

plt.xlabel('% of Weights pruned')
plt.ylabel('Accuracy')
plt.title('EXP2_VGG_CIFAR10')
plt.legend()
plt.grid(True)
plt.savefig("Exp2_VGG_cifar10.pdf")
```

```python

df2 = df2.rename(columns = {"Unnamed: 1":"% Pruned", "Accuracy on CIFAR100":"CIFAR10", "Unnamed: 3":"CIFAR100", "Unnamed: 4":"Random"})
df2 = df2.iloc[1:]
df2.dropna(inplace = True)
# Select specific columns for plotting
idx_vals = np.float64(df2["% Pruned"].values)
x_values = idx_vals
y_values = np.float64(df2["CIFAR10"].values)

# Plotting the graph
plt.plot(x_values, y_values,label='Cifar 10', marker = '.')
y_values = np.float64(df2["CIFAR100"].values)
plt.plot(x_values, y_values,label='Cifar 100', marker = '^')
y_values = np.float64(df2["Random"].values)
plt.plot(x_values, y_values,label='Random', marker = '*')

plt.xlabel('% of Weights pruned')
plt.ylabel('Accuracy')
plt.title('EXP2_VGG_CIFAR100')
plt.legend()
plt.grid(True)
plt.savefig("Exp2_VGG_cifar100.pdf")
```

```python
df3 = df3.rename(columns = {"Unnamed: 1":"% Pruned", "Accuracy on SVHN":"CIFAR10", "Unnamed: 3":"CIFAR100", "Unnamed: 4":"SVHN","Unnamed: 5":"Random"})
df3 = df3.iloc[1:]
df3.dropna(inplace = True)
# Select specific columns for plotting
idx_vals = np.float64(df3["% Pruned"].values)
x_values = idx_vals
y_values = np.float64(df3["CIFAR10"].values)

# Plotting the graph
plt.plot(x_values, y_values,label='Cifar 10', marker = '.')
y_values = np.float64(df3["CIFAR100"].values)
plt.plot(x_values, y_values,label='Cifar 100', marker = '^')
y_values = np.float64(df3["Random"].values)
plt.plot(x_values, y_values,label='Random', marker = '*')
y_values = np.float64(df3["SVHN"].values)
plt.plot(x_values, y_values,label='SVHN', marker = 'x')


plt.xlabel('% of Weights pruned')
plt.ylabel('Accuracy')
plt.title('EXP2_VGG_CIFAR100')
plt.legend()
plt.grid(True)
plt.savefig("Exp2_VGG_SVHN.pdf")
```
