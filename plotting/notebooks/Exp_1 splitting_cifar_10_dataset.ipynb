```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load data from Excel file
xls = pd.ExcelFile("/kaggle/input/final-plots/Experiment - 1 - VGG.xlsx")
```

```python
df1 = pd.read_excel(xls, sheet_name="VGG", index_col=0)
```

```python
df1 = df1.rename(columns = {"Unnamed: 1":"% Pruned", "Accuracy on CIFAR10-B":"CIFAR-10A", "Unnamed: 3":"Random", "Unnamed: 4":"CIFAR10-B"})
df1 = df1.iloc[1:]
df1.dropna(inplace = True)
# Select specific columns for plotting
idx_vals = np.float64(df1["% Pruned"].values)
x_values = idx_vals
y_values = np.float64(df1["CIFAR-10A"].values)

# Plotting the graph
plt.plot(x_values, y_values,label='Cifar 10a', marker = '.')
y_values = np.float64(df1["CIFAR10-B"].values)
plt.plot(x_values, y_values,label='Cifar 10b', marker = '^')
y_values = np.float64(df1["Random"].values)
plt.plot(x_values, y_values,label='Random', marker = '*')

plt.xlabel('% of Weights pruned')
plt.ylabel('Accuracy')
plt.title('EXP1_VGG')
plt.legend()
plt.grid(True)
plt.savefig("Exp1_VGG_N.pdf")
```
