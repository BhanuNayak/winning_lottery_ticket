```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load data from Excel file
xls = pd.ExcelFile("/kaggle/input/final-plots/Experiment - 3.xlsx")
```

```python
df1 = pd.read_excel(xls, sheet_name="VGG", index_col=0)
df1 = df1.rename(columns = {"Unnamed: 1":"% Pruned", "Accuracy on CIFAR10":"Adam after SGD", "Unnamed: 3":"SGD after Adam", "Unnamed: 4":"SGD after SGD", "Unnamed: 5":"Adam after Adam", "Unnamed: 6":"Random(SGD)", "Unnamed: 7":"Random(Adam)"})
df1 = df1.iloc[1:]
df1.dropna(inplace = True)
# Select specific columns for plotting
idx_vals = np.float64(df1["% Pruned"].values)
x_values = idx_vals
y_values = np.float64(df1["SGD after Adam"].values)

# Plotting the graph
plt.plot(x_values, y_values,label='Adam', marker = '.')
y_values = np.float64(df1["SGD after SGD"].values)
plt.plot(x_values, y_values,label='SGD', marker = '^')
y_values = np.float64(df1["Random(SGD)"].values)
plt.plot(x_values, y_values,label='Random', marker = '*')

plt.xlabel('% of Weights pruned')
plt.ylabel('Accuracy')
plt.title('EXP3_VGG')
plt.legend()
plt.grid(True)
plt.savefig("Exp3_VGG_SGD.pdf")
```
